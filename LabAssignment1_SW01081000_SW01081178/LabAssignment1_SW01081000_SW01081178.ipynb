```python
# importing libraries

import re # import regular expression module
import json  # import the JSON module
import requests
import csv # import CSV module for reading and writing CSV files
from datetime import datetime # import datetime class from datetime module
```

```python
# Extracting ID from the URL using regular expressions

# Extracting shop ID and item ID from Shopee URL using regular expressions
url = 'https://shopee.com.my/Apple-iPad-(10th-generation)-Wi-Fi-only-i.304504082.22305390280?sp_atk=449e4051-dce0-4587-8c63-88fca24c65b0&xptdk=449e4051-dce0-4587-8c63-88fca24c65b0'

r = re.search(r'i\.(\d+)\.(\d+)', url) # Check if the regular expressions pattern matched in the URL
if r:
    shop_id, item_id = r.group(1), r.group(2) # If pattern didn't match, print an error message and exit
else:
    print("Invalid URL")
    exit()
```

```python
# save reviews to CSV file

with open('shopee_review.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Username', 'Date', 'Comment']) # initialize a CSV writer with specified field names
    writer.writeheader() # write header row to the CSV file
    
    offset = 0
    page = 3
    while page > 0:    # loop until desired number of pages 
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
        data = requests.get(ratings_url).json()

        for rating in data['data']['ratings']:  # iterate through each rating in the response data
            writer.writerow({'Username': rating['author_username'],  # write the username, date, and comment to the CSV file
                             'Date': datetime.fromtimestamp(rating['ctime']).strftime('%Y-%m-%d'), 
                             'Comment': rating['comment']})
            
            print(rating['author_username'])  # print username and comment for each rating 
            print(rating['comment'])
            print('-' * 80)

        if len(data['data']['ratings']) < 20:   # check if the number of ratings received is less than 20
            break

        offset += 20
        page -= 1

print("Reviews have been saved to shopee_review.csv") 
# print a message indicating that the reviews have been saved to the CSV file
```
